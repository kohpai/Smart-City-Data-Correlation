```python
import seaborn as sns

# Load one of the data sets that come with seaborn
tips = sns.load_dataset("tips")
```

```python
print(tips)
```

```python
sns.jointplot("total_bill", "tip", tips, kind='reg')
```

```python
sns.lmplot("total_bill", "tip", tips, col="smoker")
```
