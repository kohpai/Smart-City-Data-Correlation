```python
# Import the `pandas` library as `pd`
import pandas as pd
```

```python
# Load in the data with `read_csv()`
iris = pd.read_csv("data/Iris/iris.data", header=None,
                     names=["sepal_length", "sepal_width", "petal_length", "petal_width", "breed"])

digits = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra",
                     header=None)
```

```python
# Petal length greater than sepal length?
iris.query('petal_length > sepal_length')
```

```python
# Petal length equals sepal length?
iris.query('petal_length == sepal_length')
```

```python
# Another way to query
iris[iris.petal_length > iris.sepal_length]
```

```python
# Identify missing values
pd.isnull(iris)
```

```python
# Factorize the values
levels,labels = pd.factorize(iris.breed)

# Save the encoded variables in `iris.breed`
iris.breed = levels

# Print out the first row
iris.breed.head(1)
```
